```python
#Importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#Importing dataset
df=pd.read_csv("/content/drive/MyDrive/weatherAUS.csv.zip")
```

```python
#Dropping Date column from table.
df.drop(["Date"],axis=1,inplace=True)
```

```python
#Exploring numeric feature from dataframe
numeric_feature=df.describe(include=float).columns
numeric_feature
```

```python
#Exploring all categorical_feature from dataframe.
categorical_feature=df.describe(include=object).columns
categorical_feature
```

```python
#Creating new dataframe by cloning df dataframe.
df1=df.copy()
```

```python
#Exploring all null values from target feature.
a=[(i,index) for i,index in enumerate(df["RainTomorrow"]) if pd.isna(index)]
result = pd.DataFrame(a).set_index(0)
```

**Extracting inferencial data**

```python
#Converting list of nan value target variable into dataframe.
t=pd.DataFrame(a).set_index(0)
t.head()
```

```python
t.shape
```

```python
index_nan=t.index.tolist()
index_nan
```

```python
#Exploring the inference dataset.
df1_inference=df1.iloc[index_nan]
df1_inference
```

```python
#Creating list of all non null values in the target variable for spliting in train and test data.
a=[]
for i in range(df1.shape[0]):
  if i not in index_nan:
    a.append(i)
```

**Separating training data**

```python
df_train_data=df.iloc[a]
df_train_data
```

**Missing values tratment for known data**

```python
df_train_data=df_train_data.fillna(method="ffill",axis=0).fillna(method="bfill",axis=0)
df_train_data
```

**known data splitting for further scaling. i am not appling the scaling on the target feature**

```python
x_training=df_train_data.drop(["RainTomorrow"],axis=1)
y_training=df_train_data["RainTomorrow"]
```

**lebal encoding on x_known data**

```python
#Label encoding on known data
from sklearn import preprocessing
L_Encoder=preprocessing.LabelEncoder()
x_training["Location"]=L_Encoder.fit_transform(x_training["Location"])
```

```python
x_training["Location"].unique()
```

```python
x_training["WindGustDir"]=L_Encoder.fit_transform(x_training["WindGustDir"])
```

```python
x_training["WindDir9am"]=L_Encoder.fit_transform(x_training["WindDir9am"])
```

```python
x_training["WindDir3pm"]=L_Encoder.fit_transform(x_training["WindDir3pm"])
```

```python
x_training["RainToday"]=L_Encoder.fit_transform(x_training["RainToday"])
```

```python
x_training.head()
```

**scaling on x_known data**

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
model=scalar.fit(x_training)
x_training1=model.transform(x_training)
print(x_training1)
```

```python
x_training2=pd.DataFrame(x_training1)
```

```python
#splitting data into train test 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x_training2,y_training,random_state=0,test_size=0.20)
```

```python
#use of DecisionTreeClassifier for classification task 
from sklearn.tree import DecisionTreeClassifier 
model = DecisionTreeClassifier() 
decision_tree=model.fit (X_train, y_train) 
predictions_train = model.predict(X_train)
```

```python
#Measure the performance of the decision tree model
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
print(classification_report(y_train, predictions_train))
print(confusion_matrix(y_train, predictions_train))
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_train,predictions_train)
score
```

```python
predictions_test = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,predictions_test)
score
```
